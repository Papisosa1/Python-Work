```python
# Calculate the total number of passengers
total_passengers = len(df)

# Calculate the total number of survivors
total_survivors = df['Survived'].sum()

# Calculate the overall survival rate
survival_rate = total_survivors / total_passengers

# Calculate the percentage of passengers who survived
survival_percentage = survival_rate * 100

print(survival_percentage)
```

```python
# Calculate the number of survivors in each class
survivors_by_class = df.groupby('Pclass')['Survived'].sum()

print(survivors_by_class)
```

```python
passengers_by_class = df['Pclass'].value_counts()
survival_rate_by_class = survivors_by_class / passengers_by_class
survival_rate_by_class 
```

```python
# Calculate the total number of passengers for each gender
passengers_by_gender = df['Sex'].value_counts()

# Calculate the number of survivors for each gender
survivors_by_gender = df.groupby('Sex')['Survived'].sum()

# Calculate the survival rate for each gender
survival_rate_by_gender = survivors_by_gender / passengers_by_gender
survival_rate_by_gender
```
